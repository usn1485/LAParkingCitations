```python
import pandas as pd
import numpy as np
import gmaps
import requests
from config import gkey
from pprint import pprint
```

```python
gmaps.configure(api_key=gkey)
```

```python
small_parking = pd.read_csv('los-angeles-parking-citations/small_parking-citations.csv')
```

```python
# View dataframe to see what we're working with
small_parking.head()
```

```python
 # Get the address from dataframe
address = small_parking['Location']
    
target_url = ('https://maps.googleapis.com/maps/api/geocode/json?'
    'address={0}&key={1}').format(address, gkey)
    
response = requests.get(target_url).json()
pprint(response)
```
